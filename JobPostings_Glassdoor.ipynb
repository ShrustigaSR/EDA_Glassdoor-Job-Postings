```python
def job_simp(x):
    if 'Data Scientist' in x:
        return 'data scientist'
    elif 'Analyst' in x:
        return 'analyst'
    elif 'Data Engineer' in x:
        return 'data engineer'
    elif 'Director' in x:
        return 'director'
    elif 'Manager' in x:
        return 'manager'
    elif 'Machine Learning' in x:
        return 'mle'
    else:
        return 'na'      
```

```python
def seniority(x):
    
    x = x.lower()
    
    if 'jr' in x or 'sunior' in x:
        return 'jr'
    elif 'senior' in x or 'sr' in x or 'lead' in x  or 'principal' in x:
        return 'senior'
    else:
        return 'na'
```

```python
df_start['seniority'] = df_start['Job Title'].apply(lambda x: seniority(x))
```

Exploratory Data Analysis

```python
fig = px.choropleth(height = 600, width = 600,
    locations=df_start['job_state'].value_counts().index, # Spatial coordinates
    
    locationmode = 'USA-states', # set of locations match entries in locations
    color = df_start['job_state'].value_counts(),
    color_continuous_scale=px.colors.sequential.YlOrRd,
    labels= {'color': 'Jobs Frequency'},
    title= 'Jobs per state'
                   )
fig.update_layout(
    geo_scope='usa', # limite map scope to USA,
)
fig.show()
```

Considering the heatmap above, we can see that California has the most jobs offer and pays more than any other state as expected.


```python
fig = px.choropleth(height = 600, width = 600,
    locations= df_start.groupby('job_state')['avg_salary'].mean().index, # Spatial coordinates
    locationmode = 'USA-states', # set of locations match entries in locations
    color = df_start.groupby('job_state')['avg_salary'].mean(),
    color_continuous_scale=px.colors.sequential.YlOrRd,
    labels= {'color':'K$/Year'},
    title= 'Average Salary per state'
                   )
fig.update_layout(
    geo_scope='usa', # limite map scope to USA,
)
fig.show()
```
